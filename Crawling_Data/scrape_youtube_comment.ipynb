#### Extract Comments From Each YouTube Video

```python
def extract_comment_from_youtube_video(video_dict):
  '''
    Input : Dictionary with Title and link of each youtube video
    Extract : extract comment using youtube_comment_downloader libraries then extract emojis information of each comment using emojis libraries and filter out comments with emojis finally corresponding to each emojis find the meaning of each by using demojis  
    Output : Dictionary contains Video's Link, Video's Title, List of Comments, List of Comments with emoji, List of Emojis with the meaning of each emoji used
  '''
  demoji.download_codes()
  comment_dict = {}
  i = 1
  downloader = YoutubeCommentDownloader()
  for video in video_dict:
    comments = downloader.get_comments_from_url(video_dict[video], sort_by=SORT_BY_POPULAR)
    comment_list = list()
    comment_with_emoji_list = list()
    emojis_list = list()
    demojis_list = list()
    for comment in islice(comments, 5000):
      text = str(comment['text'])
      comment_list.append(text)
      emojis = re.findall(r'[^\w\⁠s,. ]', text)
      # emojis_list.append(emojis)
      demojis = demoji.findall(text)
      demojis_list.append(demojis)
      if len(emojis) > 0:
        comment_with_emoji_list.append(text)
    # comment_dict[i] = {'Link':video_dict[video],'Title':video, 'Comments': comment_list, "Comments_with_emoji":comment_with_emoji_list, 'Emojis': emojis_list, 'Demojis': demojis_list}
    comment_dict[i] = {'Link':video_dict[video],'Title':video, 'Comments': comment_list, "Comments_with_emoji":comment_with_emoji_list, 'Emojis': demojis_list}
    i += 1
    return comment_dict
```

#### Save Dictionary to Json File

```python
comment_dict = extract_comment_from_youtube_video(video_dict)
json_object = json.dumps(comment_dict, indent=4)
 
with open("comments.json", "w") as outfile:
    outfile.write(json_object)
```
